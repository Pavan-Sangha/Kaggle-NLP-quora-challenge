## Preprocessing our text data: Tokenisation

```python
# Tokenisation splits sentences into lists of individual words. 

def tokenize(text):
    
    tokens = re.split('\W+',text)
    return tokens
```

```python
text = 'thats >>>>>>>> pretty coooooooool'
tokenize(text)
```

```python
# note we use x.lower() to get all words on the same level and make them lower case
```

### Lemmatising

### Count vectorisation 
* Counts the number of times a given word appears in the document. 

```python
# we reload the data and apply a single transformation function
data1 = df.copy()
```

```python
data1.head()
```

```python
def clean_text(text):
    
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [ps.stem(word) for word in tokens if word not in stopwords]
    return text
```

### Apply countvectoriser 

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# we create a count vectorizer object and use clean_text as the function 

count_vect = CountVectorizer(analyzer=clean_text)
X_counts = count_vect.fit_transform(data['question_text'])
```

## N-grams

Creates a document term matrix where counts still occupy the cell but instead of the columns representing single terms, they represent all combinations of adjacent words of length n

```python
# we need to pass in a string which has been cleaned, therefore we need to modify our clean text function 
#to return a string rather than a list

def clean_text_ngram(text):
    
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = " ".join([ps.stem(word) for word in tokens if word not in stopwords])
    return text
    
    

```

```python
data1.head()
```

```python
data1['cleaned'] = data1['question_text'].apply(lambda x:clean_text_ngram(x))
```

```python
data1.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

#CountVectorizer ngram_range(1:n) means search for uni-grams, bi-grams, ... , n-grams

n_gram_vect = CountVectorizer(ngram_range = (2,2))  #only bi-grams
X_counts = n_gram_vect.fit_transform(data1['cleaned'])
```

```python
def clean_text_tfidf(text):
    
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [ps.stem(word) for word in tokens if word not in stopwords]
    return text
```
