## Preprocessing our text data: Removing punctuation

```python
pd.set_option('display.max_colwidth',100) ## set a maximum limit on the width a dataFrame shows when its printed. 
```

```python
import string
string.punctuation
```

```python
def remove_punct(text):
    
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    
    return text_nopunct
```

```python
text = "a will be'd ! not be doing anything like?"
```

```python
remove_punct(text)
```

### applying to dataset

```python
data['question_text_cleaned'] = data['question_text'].apply(lambda x: remove_punct(x))
```

```python
data.head()
```

## Preprocessing our text data: Tokenisation

```python
# Tokenisation splits sentences into lists of individual words. 

def tokenize(text):
    
    tokens = re.split('\W+',text)
    return tokens
```

```python
text = 'thats >>>>>>>> pretty coooooooool'
tokenize(text)
```

```python
# note we use x.lower() to get all words on the same level and make them lower case
```

```python
data['question_text_tokenized'] = data['question_text_cleaned'].apply(lambda x: tokenize(x.lower()))
```

```python
data
```

### removing stopwards

```python
### Remove stop words

stopwords = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(tokenized_list):
    
    text = [word for word in tokenized_list if word not in stopwords]
    
    return text
    
    
```

```python
data['question_text_nonstop'] = data['question_text_tokenized'].apply(lambda x: remove_stopwords(x))
```

## preprocessing: Stemming

```python
# reducing the number of words by reducing variations of the same root word to the root
# stemming,stemmed --> stem     berry, berries --> beri     
# a case where its wrong meaning, meaness --> mean  (not correct)

```

### Porter Stemmer

```python
ps = nltk.PorterStemmer()
```

```python
dir(ps)
```

```python
print(ps.stem('grows'))
print(ps.stem('growing'))
print(ps.stem('grow'))
```

```python
# first two are actions the last one describes a person
print(ps.stem('run'))
print(ps.stem('running'))
print(ps.stem('runner'))
```

```python
def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text
```

```python
data.head()
```

```python
data['body_text_stemmed'] = data['question_text_nonstop'].apply(lambda x: stemming(x))
```

```python
data.head()
```

### Lemmatising

```python
# lemmatising is similar to stemming but lemmatisers always return a dictionary word
```

```python
wn = nltk.WordNetLemmatizer()
```

```python
# lemmatizing can tell the difference between meanness and meaning
print(wn.lemmatize('meanness'))
print(wn.lemmatize('meaning'))
```

```python
print(ps.stem('goose'))
print(ps.stem('geese'))

# the porter stemmer doesn't really know what to do
```

```python
# the lemmatizer however can tell they have the same root

print(wn.lemmatize('goose'))
print(wn.lemmatize('geese'))

```

```python
def lemmatizing(tokenized_text):
    
    text = [wn.lemmatize(word) for word in tokenized_text]
    return text
```

```python
data.head()
```

```python
data['body_text_lemmatized'] = data['question_text_nonstop'].apply(lambda x: lemmatizing(x))
```
